```python
import random

personas = []

def mostrar_menu():
    print("Bienvenido al Registro de Ciudadanos de la Unión Europea")
    print("Seleccione una opción:")
    print("1. Grabar")
    print("2. Buscar")
    print("3. Imprimir certificados")
    print("4. Salir")

def grabar_persona():
    nif = input("Ingrese el NIF de la persona (formato: 03034567-XXY): ")
    nombre = input("Ingrese el nombre de la persona (mínimo 8 caracteres): ")
    edad = int(input("Ingrese la edad de la persona: "))
    pertenece_ue = input("¿Pertenece a la Unión Europea? (Sí/No): ")

    if not validar_nif(nif):
        print("El formato del NIF es incorrecto.")
        return

    if len(nombre) < 8:
        print("El nombre debe tener al menos 8 caracteres.")
        return

    if edad < 0:
        print("La edad debe ser mayor o igual a 0.")
        return

    pertenece_ue = pertenece_ue.lower()
    if pertenece_ue == "sí" or pertenece_ue == "si":
        pertenece_ue = True
    elif pertenece_ue == "no":
        pertenece_ue = False
    else:
        print("Respuesta inválida para la pertenencia a la Unión Europea. Por favor, ingrese 'Sí' o 'No'.")
        return

    personas.append({"NIF": nif, "Nombre": nombre, "Edad": edad, "Pertenece UE": pertenece_ue})
    print("La persona ha sido registrada correctamente.")

def buscar_persona():
    nif = input("Ingrese el NIF de la persona que desea buscar: ")

    for persona in personas:
        if persona["NIF"] == nif:
            print("Información de la persona:")
            print(f"NIF: {persona['NIF']}")
            print(f"Nombre: {persona['Nombre']}")
            print(f"Edad: {persona['Edad']}")


            if persona["Pertenece UE"]:
                print("Pertenece a la Unión Europea.")
            else:
                print("No pertenece a la Unión Europea.")

            return

    print("La persona no ha sido encontrada.")

def imprimir_certificados():
    certificado = input("Ingrese el certificado que desea imprimir (Nacimiento/Conyuge/UnionEuropea): ")

    if certificado.lower() == "nacimiento":
        nombre = input("Ingrese el nombre de la persona: ")
        fechanac = input("Ingrese la fecha de nacimiento de la persona: ")
        print("------------------------")
        print("Certificado de Nacimiento")
        print(f"Nombre: {nombre}")
        print(f"Fecha de Nacimiento: {fechanac}")
        print("------------------------")
    elif certificado.lower() == "conyuge":
        nombre = input("Ingrese el nombre de la persona: ")
        estado_conyugal = input("Ingrese el estado conyugal de la persona: ")
        print("------------------------")
        print("Certificado de Estado Conyugal")
        print(f"Nombre: {nombre}")
        print(f"Estado Conyugal: {estado_conyugal}")
        print("------------------------")
    elif certificado.lower() == "unioneuropea":
        nombre = input("Ingrese el nombre de la persona: ")
        pertenece_ue = input("¿Pertenece a la Unión Europea? (Sí/No): ")
        print("------------------------")
        print("Certificado de Pertenencia a la Unión Europea")
        print(f"Nombre: {nombre}")
        print(f"Pertenece a la Unión Europea: {pertenece_ue}")
        print("------------------------")
    else:
        print("Certificado inválido.")
        return

def validar_nif(nif):

    if len(nif) != 12:
        return False
    if not nif[:8].isdigit():
        return False
    if nif[8] != "-":
        return False
    if not nif[9:].isalpha() or len(nif[9:]) != 3:
        return False

    return True

def pertenece_union_europea(nif):

    return nif[:2] == "99"

mostrar_menu()

while True:
    opcion = input("Ingrese el número de opción que desea ejecutar: ")

    if opcion == "1":
        grabar_persona()
    elif opcion == "2":
        buscar_persona()
    elif opcion == "3":
        imprimir_certificados()
    elif opcion == "4":
        print("¡Gracias por utilizar el programa!")
        print("Autor: [Tu nombre y apellido]")
        print("Versión: 1.0")
        break
    else:
        print("Opción inválida. Por favor, ingrese un número válido.")

    print("------------------------")
    mostrar_menu()
```
